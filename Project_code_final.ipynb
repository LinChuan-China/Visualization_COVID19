# The replication of the original graph

```python
import requests
import json
import re
from pyecharts.charts import Map
from pyecharts import options
#发起网络请求，获取数据 Send web request, getting data
result = requests.get('https://interface.sina.cn/news/wap/fymap2020_data.d.json?1580097300739&&callback=sinajp_1580097300873005379567841634181')
#使用正则表达式处理数据 deal the data with Regular expression
json_str = re.search("\(+([^)]*)\)+", result.text).group(1)
html = f"{json_str}"
table = json.loads(f"{html}")
title="The accumulative number of cases of COVID-19 for each province for NOW\n Which province has more cases?"
province_data = []
for province in table['data']['list']:
    province_data.append((province['name'], province['value']))
    city_data = []
    for city in province['city']:
        city_data.append((city['mapName'], city['conNum']))

map_country = Map()
map_country.set_global_opts(title_opts=options.TitleOpts(title=title), visualmap_opts=options.VisualMapOpts(is_piecewise=True,#设置是否为分段显示
                pieces=[
                    {"min": 1000, "label": '>1000 people', "color": "#6F171F"},
                    {"min": 500, "max": 1000, "label": '500-1000 people', "color": "#C92C34"},
                    {"min": 100, "max": 499, "label": '100-499 people', "color": "#E35B52"},
                    {"min": 10, "max": 99, "label": '10-99 people', "color": "#F39E86"},
                    {"min": 1, "max": 9, "label": '1-9 people', "color": "#FDEBD0"}]))
map_country.add("",province_data, maptype="china")
map_country.render("country.html")
map_country.render_notebook()
```

### Tips:

### 2. This graph is always based on the newest data, which is the data of the day when you run this code. The reason is that the data is got by crawling the web.
### 3. The characters within the graph are province names.

### assume that we have counrty CC, which have province A,B,C,D,E,F

```python
import matplotlib.pyplot as plt
import numpy as np
def get_xs(contour):
    return [ point[0] for point in contour]

def get_ys(contour):
    return [ point[1] for point in contour]
def plot_c(contour):
    plt.plot(get_xs(contour),get_ys(contour),color="black")

def plot_c2(contour,string):
    plt.plot(get_xs(contour),get_ys(contour),color="black")
    x=np.mean(get_xs(contour))
    y=np.mean(get_ys(contour))
    plt.text(s=string,x=x,y=y)
plt.figure(figsize=(10,8))
contour_A=[(2,1),(6,1),(3.5,2),(2.5,2.5),(1,3),(2,1)]
contour_B=[(1,3),(2.5,2.5),(3,4),(3.7,4),(4,5),(2,6),(1,3)]
contour_C=[(2.5, 2.5), (3.5, 2), (4, 2.5), (3.7, 4), (3, 4), (2.5, 2.5)]
contour_D=[(3.5, 2), (6, 1), (9, 5), (4, 2.5), (3.5, 2)]
contour_E=[(4, 2.5), (9, 5), (5, 10), (4, 5), (3.7, 4)]
contour_F=[(4, 5), (5, 10), (2, 6), (4, 5)]
plot_c2(contour_A,"A")
plot_c2(contour_B,"B")
plot_c2(contour_C,"C")
plot_c2(contour_D,"D")
plot_c2(contour_E,"E")
plot_c2(contour_F,"F")
plt.title("country CC")
```

### Assume that in the last 30 days, the number of new cases of COVID-19 for province A,B,C,D,E,F are 10, 8, 2, 7, 11, 3 respectively. What is more, we know their locations.

```python
cases_A=[(3.5,1.9),(3.4,2),(3.4,1.8),(3.2,1.8),(3.3,1.7),(3.2,1.5),(3.8,1.3),(3.7,1.4),(3.6,1.5),(3.6,1.65)]
cases_B=[(3.9,5),(3.5,4.9),(3.4,4.5),(3.7,4.1),(3.6,4.2),(3.5,5),(3.6,5.1),(3.7,4.5)]
cases_C=[(3.5,2.2),(3.6,3.8)]
cases_D=[(3.6,2.1),(3.7,2.15),(3.7,1.85),(4,2),(4,2.1),(3.9,2.15),(3.95,2.2)]
cases_E=[(4,4.1),(4,4.2),(4,4.3),(4.1,4.3),(4.5,4.4),(4.5,3.9),(3.85,4),(3.9,3.5),(4.5,6),(4.4,5.2),(4.5,4.3)]
cases_F=[(4,6),(4,5.5),(3.8,5.5)]

def get_alpha(n):
    n=len(n)
    if n>=10:
        return 0.5
    elif n>=5:
        return 0.3
    elif n>0:
        return 0.1
    else:
        return 0

def plot_cases1(contour,cases):
    alpha=get_alpha(cases)
    plt.fill(get_xs(contour),get_ys(contour),alpha=alpha,color="red")
    
def plot_cases2(cases):
    plt.plot(get_xs(cases),get_ys(cases),"ro",markersize=3)

plt.figure(figsize=(10,8))
plot_c2(contour_A,"A")
plot_c2(contour_B,"B")
plot_c2(contour_C,"C")
plot_c2(contour_D,"D")
plot_c2(contour_E,"E")
plot_c2(contour_F,"F")
plot_cases1(contour_A,cases_A)
plot_cases1(contour_B,cases_B)
plot_cases1(contour_C,cases_C)
plot_cases1(contour_D,cases_D)
plot_cases1(contour_E,cases_E)
plot_cases1(contour_F,cases_F)
plt.title("The graph based on the original principles")
```
